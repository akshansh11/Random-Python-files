```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load the image
image = cv2.imread("pisa.jpg")

# Convert the image from BGR to RGB (OpenCV uses BGR by default)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Split the RGB channels
red_channel, green_channel, blue_channel = cv2.split(image)

# Create a blank image with the same size as the input image for visualization
blank_image = np.zeros_like(red_channel)

# Merge the individual channels with blank images to create a visual representation
red_image = cv2.merge([red_channel, blank_image, blank_image])
green_image = cv2.merge([blank_image, green_channel, blank_image])
blue_image = cv2.merge([blank_image, blank_image, blue_channel])

# Visualize the original image and the individual RGB channels
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
axes[0].imshow(image)
axes[0].set_title("Original Image")
axes[1].imshow(red_image)
axes[1].set_title("Red Channel")
axes[2].imshow(green_image)
axes[2].set_title("Green Channel")
axes[3].imshow(blue_image)
axes[3].set_title("Blue Channel")

# Remove axis ticks for a cleaner look
for ax in axes:
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("pisa.jpg")

# Convert the image from BGR to RGB (OpenCV uses BGR by default)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Method 1: Averaging the RGB channels
average_gray = np.mean(image, axis=2).astype(np.uint8)

# Method 2: Using a weighted sum based on each color's perceived luminance
luminance_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Method 3: Employing alternative color space conversions

# Convert the image to HSL color space and extract the Lightness channel
hsl_image = cv2.cvtColor(image, cv2.COLOR_RGB2HLS)
lightness_gray = hsl_image[:, :, 1]

# Convert the image to HSV color space and extract the Value channel
hsv_image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
value_gray = hsv_image[:, :, 2]

# Visualize the original image and the grayscale images using different methods
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes[0, 0].imshow(image)
axes[0, 0].set_title("Original Image")
axes[0, 1].imshow(average_gray, cmap="gray")
axes[0, 1].set_title("Average Gray")
axes[0, 2].imshow(luminance_gray, cmap="gray")
axes[0, 2].set_title("Luminance Gray")
axes[1, 0].axis("off")
axes[1, 1].imshow(lightness_gray, cmap="gray")
axes[1, 1].set_title("Lightness Gray (HSL)")
axes[1, 2].imshow(value_gray, cmap="gray")
axes[1, 2].set_title("Value Gray (HSV)")

# Remove axis ticks for a cleaner look
for ax in axes.flatten():
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("pisa.jpg")

# Convert the image to grayscale
grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Set the global threshold value
threshold_value = 128

# Apply the threshold to the image
_, binary_image = cv2.threshold(grayscale_image, threshold_value, 255, cv2.THRESH_BINARY)

# Visualize the original image, grayscale image, and binary image
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title("Original Image")
axes[1].imshow(grayscale_image, cmap="gray")
axes[1].set_title("Grayscale Image")
axes[2].imshow(binary_image, cmap="gray")
axes[2].set_title("Binary Image (Global Thresholding)")

# Remove axis ticks for a cleaner look
for ax in axes:
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("pisa.jpg")

# Convert the image to grayscale
grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to reduce noise
blurred_image = cv2.GaussianBlur(grayscale_image, (5, 5), 0)

# Apply the Canny Edge Detection algorithm
edges = cv2.Canny(blurred_image, 50, 150)

# Visualize the original image and the edges
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title("Original Image")
axes[1].imshow(edges, cmap="gray")
axes[1].set_title("Edges (Canny Algorithm)")

# Remove axis ticks for a cleaner look
for ax in axes:
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread("pisa.jpg")

# Convert the image to grayscale
grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Calculate the histogram of pixel intensities
histogram = cv2.calcHist([grayscale_image], [0], None, [256], [0, 256])

# Normalize the histogram
histogram /= histogram.sum()

# Plot the grayscale image and its histogram
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].imshow(grayscale_image, cmap="gray")
axes[0].set_title("Grayscale Image")
axes[1].bar(np.arange(256), histogram[:, 0], color="gray", alpha=0.8)
axes[1].set_title("Histogram of Pixel Intensities")
axes[1].set_xlabel("Intensity Value")
axes[1].set_ylabel("Normalized Frequency")

plt.show()
```
